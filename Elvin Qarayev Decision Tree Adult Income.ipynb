```python
# -----------------------------------------------------------------
# Decision Tree Classifier
# Predict the income of an adult based on the census data
# -----------------------------------------------------------------

# Import libraries
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns 
from sklearn.model_selection import train_test_split 
from sklearn import metrics
%matplotlib inline
```

```python
# Read dataset
adult=pd.read_csv('decisiontreeAdultIncome.csv')
adult.head()
```

```python
# Check for Null values
adult.isnull().sum()
```

```python
# Create Dummy variables
wc = pd.get_dummies(adult['wc'],drop_first=True)
education = pd.get_dummies(adult['education'],drop_first=True)
maritalstatus = pd.get_dummies(adult['marital status'],drop_first=True)
race = pd.get_dummies(adult['race'],drop_first=True)
gender = pd.get_dummies(adult['gender'],drop_first=True)
IncomeClass = pd.get_dummies(adult['IncomeClass'],drop_first=True)
adult.drop(['wc','education','marital status', 'race','gender', 'IncomeClass'],axis=1,inplace=True)
adult = pd.concat([adult,wc,education,maritalstatus,race,gender,IncomeClass ],axis=1)
adult.head()
```

```python
# Create X and Y Variables
X=adult.drop(IncomeClass,axis=1)
X
```

```python
Y=IncomeClass
```

```python
# Split the X and Y dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,Y, test_size=0.30, random_state=101)
```

```python
# Import and train classifier
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn import metrics 
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
```

```python
# Test the model
y_pred = clf.predict(X_test)
y_pred
```

```python
# Evaluate the model
print("Dogruluq:",metrics.accuracy_score(y_test, y_pred))
```
